```python
import math
import re
import time
def MAXHEAPIFY(A,i,heap_size):
    left = 2*i+1
    right = 2*i +2
    if left <=heap_size and int(A[left].split(',')[1])>int(A[i].split(',')[1]):
        largest = left
    else:
        largest =i
    if right <=heap_size and int(A[right].split(',')[1]) > int(A[largest].split(',')[1]):
        largest = right

    if largest != i:
        A[i], A[largest] = A[largest], A[i]
        MAXHEAPIFY(A,largest,heap_size)
        
def BuildQueue(A):
    heap_size = len(A)-1
    start_point = int(math.floor(len(A)/2)-1)
    for i in range(start_point,-1,-1):
        MAXHEAPIFY(A,i,heap_size)
def HEAPSORT(A):
    print("Input: ",A)
    BuildQueue(A)
    print("Max heap",A)
    heap_size = len(A)
    for i in range(len(A)-1,0,-1):
        heap_size =heap_size-1
        A[i],A[0] = A[0],A[i] 
        MAXHEAPIFY(A,0,heap_size-1)
def Insert(A,value):
    A.append(value)
    i=len(A)-1
    while i>0 and int(A[int(i/2)].split(',')[1]) < int(value.split(',')[1]):
        A[i], A[int(i/2)]=  A[int(i/2)],A[i]
        i = int(i/2)

# Delete rule by index 
def Delete(A,index):
    if index>len(A)-1:
        print("Please select proper element")
    else:
        if index==len(A)-1:
            del A[-1]
        else:
#             swap current index to the last
            A[index], A[len(A)-1] =A[len(A)-1],A[index]
            del A[-1]
            MAXHEAPIFY(A,index,len(A)-1)
            
def ExtractMax(A):
    if len(A)<1:
        print("Heap underflow")
    else:
        max = A[0]
        last = len(A)-1
        A[0], A[last] = A[last],A[0]
        del A[-1]
        MAXHEAPIFY(A,0,len(A)-1)
    return max
def STATS(A):
    aggre_data = re.findall("\((.*?)\)", A)
    return len(aggre_data)


# Please specify test file location below this command line
filepath = 'test6.txt'
with open(filepath, 'r') as infile:

    data = infile.read()  # Read the contents of the file into memory.

# Return a list of the lines, breaking at line boundaries.
my_list = data.splitlines()
print("Reading input...%s"%filepath)
print("Basic statistics")
print("Total rules: %d"%STATS(data))
print("There are %d line(s) in our text file" %len(my_list))
print("Initialize simplified agenda manager...\n")
SAM =[]
cycle=int(1)
start_time = time.time()
for line in range(len(my_list)):
    print("\n..........................Processing CYCLE %d/30..............................................\n" %cycle)
    my_list[line] = re.findall("\((.*?)\)", my_list[line])
    if len(SAM)==0:
        print("\nBuild Priority Queue (Heap):")
        print("\nReading %d rule(s) at line %d: "%(len(my_list[line]),line+1))
        print(my_list[line])
        SAM.extend(my_list[line])
        BuildQueue(SAM)
        print("\nActivated rules in the Agenda Manager")
        print(SAM)
        print("\nExtract rule (%s) from Agenda Manager "%ExtractMax(SAM))
        print("Rebuild the priority queue....Done! %d rule(s) are in the Agenda Manager"%len(SAM))
        print(SAM)
        cycle+=1
    else:
        print("\nReading %d rule(s) at line %d: "%(len(my_list[line]),line+1))
        print(my_list[line])
        if len(my_list[line])>0:
            print("\nBuild Priority Queue (Heap):")
            for item in my_list[line]:
                print("\tAdding rule (%s) to the queue. Maintaining heap.....Done!"%item)
                Insert(SAM,item)
            print("New Agenda Manger after adding new rule(s)")
            print(SAM)
        else: 
            print("No rule added to Agenda Manager")
    
        print("\nExtract rule (%s) from Agenda Manager "%ExtractMax(SAM))
        print("Rebuild the priority queue....Done! %d rule(s) are in the Agenda Manager"%len(SAM))
        print(SAM)
        cycle+=1
        if cycle>30:
            print("\n")
            print("The system halts because it runs 30 cycles".upper())
            break

while cycle<=30 and len(SAM) >0:
    print("\n..........................Processing CYCLE %d/30..............................................\n" %cycle)
    print("Extract rule (%s) from Agenda Manager "%ExtractMax(SAM))
    print("Rebuild the priority queue....Done! %d rule(s) are in the Agenda Manager"%len(SAM))
    print(SAM)
    if cycle==30:
        print("\n")
        print("The system halts because it runs 30 cycles".upper())
        break
    if len(SAM)==0:
        print("\n")
        print("The system halts because the agenda is empty".upper())
        break
    cycle+=1
end_time=time.time()
running_time = end_time-start_time
print("Running time: %f"%running_time)
```
